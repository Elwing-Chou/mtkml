```python
from keras.datasets.mnist import load_data
(x_train, y_train), (x_test, y_test) = load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(x_train[0], cmap="gray")
```

```python
import pandas as pd
pd.DataFrame(x_train[0])
```

```python
x_train_reshape = x_train.reshape(x_train.shape[0], 784)
x_test_reshape = x_test.reshape(x_test.shape[0], 784)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=30)
x_train_pca = pca.fit_transform(x_train_reshape)
# !!! no fit_transform, only transform
x_test_pca = pca.transform(x_test_reshape)
```

```python
print("Explained:", pca.explained_variance_ratio_)
print("Sum:", sum(pca.explained_variance_ratio_))
```
