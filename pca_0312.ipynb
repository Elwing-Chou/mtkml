```
from keras.datasets.mnist import load_data
# ((90%x, 90%y), (10%x, 10%y))
(x_train, y_train), (x_test, y_test) = load_data()
```

```
%matplotlib inline
import matplotlib.pyplot as plt
print(y_train[0])
plt.imshow(x_train[0], cmap="gray")
```

```
import pandas as pd
pd.DataFrame(x_train[0])
```

```
print(x_train.shape)
print(x_test.shape)
x_train_reshape = x_train.reshape(-1, 784)
x_test_reshape = x_test.reshape(-1, 784)
pd.DataFrame(x_train_reshape)
```

```
from sklearn.decomposition import PCA
pca = PCA(n_components=30)
x_train_pca = pca.fit_transform(x_train_reshape)
x_test_pca = pca.transform(x_test_reshape)
pd.DataFrame(x_train_pca)
```

```
print("係數(30):", pca.singular_values_)
print("解釋了多少variance:", pca.explained_variance_ratio_)
print("加起來:", sum(pca.explained_variance_ratio_))
pd.DataFrame(pca.components_)
```

```
# subplot(總高度, 總寬度, 第幾張(1))
plt.figure(figsize=(14, 20))
for i, axis in enumerate(pca.components_):
    plt.subplot(6, 5, i+1)
    plt.title("Axis {}".format(i))
    plt.imshow(axis.reshape(28, 28), cmap="gray")
```

```
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
pca = PCA(n_components=2)
x_test_pca = pca.fit_transform(x_test_reshape)
scaler = MinMaxScaler()
x_test_pca_scale = scaler.fit_transform(x_test_pca)
plt.figure(figsize=(14, 14))
sns.scatterplot(x_test_pca_scale[:, 0],
                x_test_pca_scale[:, 1],
                hue=y_test,
                palette=sns.color_palette("hls", 10))
```
