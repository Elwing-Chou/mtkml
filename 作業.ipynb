```python
from keras.datasets.mnist import load_data
(x_train, y_train), (x_test, y_test) = load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
idx = random.randint(0, x_train.shape[0] - 1)
plt.title("This is " + str(y_train[idx]))
plt.imshow(x_train[idx], cmap="gray")
```

```python
import pandas as pd
pd.DataFrame(x_train[idx])
```

```python
x_train_reshape = x_train.reshape(60000,  -1)
x_test_reshape = x_test.reshape(10000,  -1)
pd.DataFrame(x_train_reshape)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=40)
x_train_pca = pca.fit_transform(x_train_reshape)
x_test_pca = pca.transform(x_test_reshape)
```

```python
pd.DataFrame(x_train_pca)
```

```python
print(":", pca.explained_variance_ratio_)
print(":", sum(pca.explained_variance_ratio_))
```
